# Data Modeling

#### Merge the data
We recommend using the Reaction table as your base table, then first join the relevant columns from your Content data set, and then the Reaction Types data set.

```python
# load clean data
contentDf = pd.read_csv('CleanContent.csv')
reactionDf = pd.read_csv('CleanReactions.csv')
reactionTypeDf = pd.read_csv('ReactionTypes.csv')
```

```python
print("Columns of Reaction dataset: ", reactionDf.columns)
print("Columns of Content dataset: ", contentDf.columns)
print("Columns of Reaction Type dataset: ", reactionTypeDf.columns)
```

```python
#merge the two datasets with reaction tabel as base table and join Content dataset then reaction type dataset
mergedDf = reactionDf.merge(contentDf, on='Content ID', how='left')
mergedDf = mergedDf.merge(reactionTypeDf, on='Type', how='left')
```

```python
mergedDf.head()
```

```python
mergedDf.columns
```

```python
mergedDf.isnull().sum()
```

## 2.  Figure out the Top 5 performing categories

```python
mergedDf['Category'].unique()
```

```python
mergedDf.groupby('Category')['Score'].sum()
```

```python
##Figure out the Top 5 performing categories based on score
# Add up the total scores for each category.
category_scores = mergedDf.groupby('Category')['Score'].sum()

# Sort the categories by their total scores in descending order.
top_categories = category_scores.sort_values(ascending=False)
```

```python
print("Top 5 cateogries are: ", top_categories.head(5))
```

```python
# Save the clean dataset of top 5 cateogries
top_categories.head(5).to_csv('Top5Categories.csv')
```

```python
mergedDf.head()
#save mergedDf
mergedDf.to_csv('CleanMergedData.csv', index=False)
```

```python
#plot the scores based on categoriesin sorted order
mergedDf.groupby('Category')['Score'].sum().sort_values(ascending=False).plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Total Score')
plt.title('Total Score by Category')
plt.show()
```

```python
#Answer some questions
# How many unique categories are there?
unique_categories = mergedDf['Category'].nunique()
print("Number of unique categories:", unique_categories)
```

```python
# How many reactions are there to the most popular category?
most_popular_category = mergedDf['Category'].value_counts().idxmax()
print("Most popular category:", most_popular_category)

#get the total number of reactions for most popular category
total_reactions = mergedDf[mergedDf['Category'] == most_popular_category]['Score'].sum()
print("Total number of reactions for most popular category:", total_reactions)
```

```python
#plot the number of posts based on months with x - axis as month name and sorted order
mergedDf.groupby('Month')['Score'].sum().sort_values(ascending=False).plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Number of Posts')
plt.title('Number of Posts by Month')
plt.show()

```

```python
#top 5 cateogries based on aggregate "popularity" score
top_5_categories = mergedDf.groupby('Category')['Score'].sum().sort_values(ascending=False).head(5)
print(top_5_categories)
```

```python
#plot the top_5_cateogries
top_5_categories.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Popularity Score')
plt.title('Top 5 Categories by Popularity Score')
plt.show()
```

```python
#popularity percentage share for top 5 cateogries
top_5_categories_percentage = top_5_categories / mergedDf['Score'].sum() * 100
print(top_5_categories_percentage)
```

```python
#create pie chart with highest percentage share poping out
plt.pie(top_5_categories_percentage, labels=top_5_categories.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Popularity Percentage Share for Top 5 Categories')
plt.show()
```
